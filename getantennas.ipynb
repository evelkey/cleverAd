```python
import pandas as pd


lista=[]

def isinbp(lon,lat):
    if (lon<47.517559 and lon >47.478452 and lat< 19.106307 and lat>19.052176):
        return True
    else:
        return False
```

```python
isinbp(47.521859, 19.027346)
```

```python
df = pd.read_csv('cell_towers.csv', delimiter=',', encoding="utf-8-sig")
```

```python
for i in range(0,len(df)):
    #print(df['lon'][i],df['lat'][i])
    if isinbp(df['lat'][i],df['lon'][i]):
        print("found")
        lista.append([df['lon'][i],df['lat'][i]])
        
    
```

```python
out=[]
for item in lista:
    out.append([item[1],item[0]])
```

```python
csvfile = "towers.csv"
import csv
done=[]

#Assuming res is a flat list
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for item in out:
        
        if item not in done:
            writer.writerow([item])
            
        done.append(item)
```
