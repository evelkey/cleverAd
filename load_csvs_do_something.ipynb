```python
import pickle
import collections
import pandas as pd
```

```python
msck_="data/msc_100k.csv"
mscw_="data/msc_weekly.csv"
crmk_="data/crm_100k.csv"
crmw_="data/crm_weekly.csv"
tack_="data/tac_100k.csv"
tacw_="data/tac_weekly.csv"
```

```python
pd.low_memory=False
msck = pd.read_table(open(msck_,encoding='utf-8'),header=None,low_memory=False,sep=';')
mscw = pd.read_table(open(mscw_,encoding='utf-8'),header=None,low_memory=False,sep=';')
crmk = pd.read_table(open(crmk_,encoding='utf-8'),header=None,low_memory=False,sep=';')
crmw = pd.read_table(open(crmw_,encoding='utf-8'),header=None,low_memory=False,sep=';')
tack = pd.read_table(open(tack_,encoding='utf-8'),header=None,low_memory=False,sep=';')
tacw = pd.read_table(open(tacw_,encoding='utf-8'),header=None,low_memory=False,sep=';')
```
